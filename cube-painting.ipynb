```python
import numpy as np
from copy import copy

def hspin(cube):
    c = copy(cube)
    c[1], c[2], c[3], c[4] = c[2], c[4], c[1], c[3]
    return c

def vspin(cube):
    c = copy(cube)
    c[0], c[1], c[4], c[5] = c[4], c[0], c[5], c[1]
    return c

def CubePainting(cubestring):
    c1, c2 = cubestring[:6], cubestring[6:]
    
    # 16 cases
    c = list(copy(c1))
    for _ in range(4):
        c = vspin(c)
        for _ in range(4):
            c = hspin(c)
            if list(c2) == c:
                return True
            
    # 4 cases
    c = list(copy(c1))
    c[0], c[2], c[3], c[5] = c[2], c[5], c[0], c[3]
    for _ in range(4):
        c = hspin(c)
        if list(c2) == c:
            return True
    
    # 4 cases
    c = list(copy(c1))
    c[0], c[2], c[3], c[5] = c[3], c[0], c[5], c[2]
    for _ in range(4):
        c = hspin(c)
        if list(c2) == c:
            return True
    
    return False
```

```python
cubestring1 = 'rbgggrrggbgr'
cubestring2 = 'rrrbbbrrbbbr'
cubestring3 = 'rbgrbgrrrrrg'

res1 = CubePainting(cubestring1)
print(res1)

res2 = CubePainting(cubestring2)
print(res2)

res3 = CubePainting(cubestring3)
print(res3)
```
